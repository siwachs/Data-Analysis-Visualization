```python
# Removing Duplicates
# Fixing NULL values(Age, Department)
# Standardization of each value(Gender)
# Concatenate First Name and Last Name
# Add a new column(YearsAtCompany)
# Strength by Maritial Status
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.shape
```

```python
df[df.duplicated()].head()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.info()
```

```python
df[df['Age'].isnull()].head()
```

```python
df[df['Department'].isnull()].head()
```

```python
averageAge = df['Age'].mean()
df['Age'] = df['Age'].fillna(int(averageAge))

df[df['Age'].isnull()]
```

```python
df['Department'] = df['Department'].fillna('Bench')
df[df['Department'].isnull()]
```

```python
df['Gender'] = df['Gender'].replace({'M' : 'Male', 'F' : 'Female'})
df.head()
```

```python
df['Full Name'] = df['First Name'] + ' ' + df['Last Name']
df.head()
```

```python
columns = df.columns.tolist()
columns
```

```python
lastNameIndex = columns.index('Last Name')
lastNameIndex
```

```python
columns.insert(lastNameIndex + 1, columns.pop(columns.index('Full Name')))
columns
```

```python
df = df[columns]
df.head()
```

```python
hireDate = pd.to_datetime(df['Hire_Date'], errors='coerce')
today = pd.to_datetime(datetime.today())

df['YearsAtCompany'] = ((today - hireDate) / pd.Timedelta(days=365)).round(1)
df.head()
```

```python
# Bar Chart

# Department Vise Employee Count
departmentCount = df['Department'].value_counts()

color = ['darkblue', 'darkgreen', 'yellow', 'purple']

plt.figure(figsize=(10, 6))
departmentCount.plot(kind = 'bar', color = color)

plt.title('Employee Count By Department', fontweight = 'bold')
plt.xlabel('Department', fontweight = 'bold')
plt.xticks(rotation = 0)
plt.ylabel('Employee Count', fontweight = 'bold')

plt.show()
```

```python
# Line Chart

ageGender = df.groupby(['Age', 'Gender']).size().reset_index(name = 'Count')
ageGender
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(data = ageGender, x = 'Age', y = 'Count', hue = 'Gender', marker = 'o')

plt.title('Employee Count By Age and Gender', fontweight = 'bold')
plt.xlabel('Age', fontweight = 'bold')
plt.xticks(rotation = 0)
plt.ylabel('Employee Count', fontweight = 'bold')

plt.show()
```

```python
# Pie Chart

maritalStatusDf = df.groupby(['MaritalStatus']).size().reset_index(name = 'Count')
maritalStatus = maritalStatusDf['MaritalStatus']
maritalStatusCount = maritalStatusDf['Count']
```

```python
sns.set_theme(style = 'whitegrid')
colors = sns.color_palette('colorblind')[0 : len(maritalStatus)]
colors = ['darkblue', 'darkgreen', 'yellow']

plt.figure(figsize=(8, 8))
plt.pie(maritalStatusCount, labels = maritalStatus, colors = colors, autopct = '%1.1f%%', startangle = 140)
plt.axis('equal')

plt.xlabel('Strenth by Marital Status', fontweight = 'bold')
plt.show()
```

```python
# Average Salary in Department By Gender

df = df.rename(columns = {'MonthlyRate' : 'Salary'})
df.head()
```

```python
averageSalary = df.groupby(['Department', 'Gender'])['Salary'].mean().reset_index()
averageSalary
```

```python
plt.figure(figsize=(16, 8))
plt.title('Average Salary by Department and Gender', fontweight = 'bold')

sns.barplot(data = averageSalary, x = 'Department', y = 'Salary', hue = 'Gender')

plt.show()
```

```python
# Histogram

age = df['Age']
age
```

```python
plt.hist(age, bins = 100, edgecolor = 'black', color = 'skyblue')

plt.title('Employee Count by Age', fontweight = 'bold')
plt.xlabel('Age', fontweight = 'bold')
plt.ylabel('Employee Count', fontweight = 'bold')
```

```python
# Box Plot Get Min, Max, Mean Median and Outliers

plt.figure(figsize = (10, 6))
plt.title('Salary by Department', fontweight = 'bold')
plt.xlabel('Department', fontweight = 'bold')
plt.ylabel('Salary', fontweight = 'bold')

sns.boxplot(x = 'Department', y = 'Salary', data = df, hue = 'Gender', showmeans = True, meanprops = {"marker" : "o",
                                                                                                        "markerfacecolor" : "white",
                                                                                                        "markersize" : "6",
                                                                                                        "markeredgecolor" : "green"})

plt.show()
```
